```python
# Define a function to reclassify land use categories
def reclassify_land_use(code):
    if code in ['101', '102', '103', '106', '109']:  # Low-Density Residential
        return 'Residential_Low'
    elif code in ['104', '105','111']:  # Medium-Density Residential
        return 'Residential_Medium'
    elif code in ['112']:  # High-Density Residential
        return 'Residential_High'
    elif code.startswith('013'):
        return 'Residential_Mixed'
    elif code.startswith('132'):
        return 'Residential_Vacant'
    elif code in ['995']:  # Open Space
        return 'Open Space'
    elif code.startswith('2'):  # Open Space
        return 'Open Space'
    elif code.startswith('3'):
        return 'Commercial'
    elif code.startswith('4'): 
        return 'Industrial'

    


    # Add more elif statements for other categories
    else:
        return 'Other'



# Apply the function to the 'land_use' column
clean['USE_CODE_RECLASSIFIED'] = clean['USE_CODE'].apply(reclassify_land_use)
```
